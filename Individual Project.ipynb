```python
import pandas as pd
import numpy as np

location = "datasets/nyc-property-sales/nyc-rolling-sales.csv"
df = pd.read_csv(location)
df.head()
```

```python
# check data types
df.dtypes
```

```python
# want to use sale price, so check for not-null entries
df[df['SALE PRICE'] != np.nan]
```

```python
df['SALE PRICE'] = pd.to_numeric(df['SALE PRICE'],errors='coerce')
# get rid of null entries of sale price
df.dropna(axis = 0, how='any', inplace=True)
```

```python
df.shape
```

```python
df['GROSS SQUARE FEET'] = pd.to_numeric(df['GROSS SQUARE FEET'],errors='coerce')
# get rid of null entries of gross square feet
df.dropna(axis = 0, how = 'any', inplace = True)
```

```python
df['GROSS SQUARE FEET'] = pd.to_numeric(df['GROSS SQUARE FEET'],errors='coerce')
# get rid of null entries of gross square feet
df.dropna(axis = 0, how = 'any', inplace = True)
```

```python
df.shape
```

```python
df['RESIDENTIAL UNITS'] = pd.to_numeric(df['RESIDENTIAL UNITS'],errors='coerce')
# get rid of null entries of residential square feet
df.dropna(axis = 0, how = 'any', inplace = True)
```

```python
df.shape
```

```python
df['TAX CLASS AT TIME OF SALE'] = pd.to_numeric(df['TAX CLASS AT TIME OF SALE'],errors='coerce')
# get rid of null entries of  tax class at time of sale
df.dropna(axis = 0, how = 'any', inplace = True)
```

```python
df.shape
```

```python
import seaborn as sns
# correlation heat map of four factors
correlation_df = df.filter(['SALE PRICE', 'GROSS SQUARE FEET', 'RESIDENTIAL UNITS', 'TAX CLASS AT TIME OF SALE'], axis = 1)
sns.heatmap(correlation_df.corr(), vmin = -1, annot = True)
```

```python
import statsmodels.formula.api as smf
```

```python
# linear regression of two factors, since by correlation heat map we can tell residential units are not very
# correlated to sale price, and neither is tax class at time of sale
regression_df = correlation_df.filter(['SALE PRICE', 'GROSS SQUARE FEET'], axis = 1)
regression_df['SALE'] = regression_df['SALE PRICE']
regression_df.head()
```

```python
regression_df['GROSS'] = regression_df['GROSS SQUARE FEET']
```

```python
result = smf.ols('SALE ~ GROSS', data=regression_df).fit()
```

```python
result.summary()
```

```python
# return to four factors to compare
correlation_df['SALE'] = correlation_df['SALE PRICE']
correlation_df['GROSS'] = correlation_df['GROSS SQUARE FEET']
correlation_df['RESIDENTIAL'] = correlation_df['RESIDENTIAL UNITS']
correlation_df['TAX'] = correlation_df['TAX CLASS AT TIME OF SALE']
result = smf.ols('SALE ~ GROSS + RESIDENTIAL + TAX', data=correlation_df).fit()
```

```python
result.summary()
```
