```python
import pandas as pd
import numpy as np

location = "datasets/nyc-property-sales/nyc-rolling-sales.csv"
df = pd.read_csv(location)
df.head()
```

```python
df[df['SALE PRICE'] != np.nan]
```

```python
origNum = df.shape
```

```python
df['SALE PRICE'] = pd.to_numeric(df['SALE PRICE'],errors='coerce')
```

```python
df['SALE PRICE']
```

```python
hasSale_df = df.dropna(axis = 0, how='any')
```

```python
hasSale_df.shape
```

```python
noDupes_df = hasSale_df.drop_duplicates()
noDupes_df.shape
```

```python
noDupes_df['GROSS SQUARE FEET'] = pd.to_numeric(noDupes_df['GROSS SQUARE FEET'],errors='coerce')
```

```python
import seaborn as sns
correlation_df = noDupes_df.filter(['SALE PRICE', 'GROSS SQUARE FEET'], axis = 1)
sns.heatmap(correlation_df.corr(), vmin = -1, annot = True)
correlation_df['GROSS SQUARE FEET'].dtypes
```

```python
noDupes_df.columns
```

```python
correlation_df = noDupes_df.filter(['SALE PRICE', 'BOROUGH'], axis = 1)
sns.boxplot(x = 'BOROUGH',y = 'SALE PRICE', data = correlation_df)
```
