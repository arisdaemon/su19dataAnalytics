```python
import numpy as np
import pandas as pd

location = "datasets/DC_Properties.csv"
df = pd.read_csv(location)
df
```

```python

df = df.drop('Unnamed: 0', axis=1)
```

```python

nodupe_df = df.drop_duplicates()
nodupe_df
```

```python
# remove all columns that are null
noNAcol_df = df.drop('ASSESSMENT_NBHD', axis = 1)
noNAcol_df = noNAcol_df.drop('CENSUS_TRACT', axis = 1)
noNAcol_df = noNAcol_df.drop('CMPLX_NUM', axis = 1)
noNAcol_df = noNAcol_df.drop('LIVING_GBA', axis = 1)
```

```python
# remove all columns that have duplicate information
nodupe_df = noNAcol_df.drop('X', axis = 1)
nodupe_df = nodupe_df.drop('Y', axis = 1)
nodupe_df = nodupe_df.drop('STORIES',axis = 1)
```

```python
nodupe_df.columns
```

```python
# remove entries that are completely null
nodupe_df = nodupe_df.dropna(axis = 0, how = 'all')
# remove entries that are missing a complete address
nodupe_df = nodupe_df.dropna(axis = 0, subset = ['FULLADDRESS','STATE'])
```

```python
# combine FULLADDRESS, CITY, STATE, ZIPCODE into a complete address
def addrCombine(full, city, state, zipc):
    return (full+" "+city+" "+state+" "+zipc)
#print(addrCombine('10 fall street','rockville','ny','1000'))
```

```python
nodupe_df['COMPLETEADDR'] = addrCombine(nodupe_df['FULLADDRESS'].astype(str), nodupe_df['CITY'].astype(str),
                                        nodupe_df['STATE'].astype(str), nodupe_df['ZIPCODE'].astype(str))
nodupe_df
```

```python
nodupe_df = nodupe_df.drop(labels=['FULLADDRESS','CITY','STATE','ZIPCODE'],axis = 1)
nodupe_df.columns
```
